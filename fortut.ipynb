```fortran
program logicalOp
! this program checks logical operators
implicit none

   ! variable declaration
   logical :: a, b
   
   ! assigning values
   a = .true.
   b = .false.
   
   if (a .and. b) then
      print *, "Line 1 - Condition is true"
   else
      print *, "Line 1 - Condition is false"
   end if
   
   if (a .or. b) then
      print *, "Line 2 - Condition is true"
   else
      print *, "Line 2 - Condition is false"
   end if
   
   ! changing values
   a = .false.
   b = .true.
   
   if (.not.(a .and. b)) then
      print *, "Line 3 - Condition is true"
   else
      print *, "Line 3 - Condition is false"
   end if
   
   if (b .neqv. a) then
      print *, "Line 4 - Condition is true"
   else
      print *, "Line 4 - Condition is false"
   end if
   
   if (b .eqv. a) then
      print *, "Line 5 - Condition is true"
   else
      print *, "Line 5 - Condition is false"
   end if
   
end program logicalOp
```

```fortran
program ifElseIfElseProg
implicit none

   ! local variable declaration
   integer :: a = 100
 
   ! check the logical condition using if statement
   if( a == 10 ) then
  
      ! if condition is true then print the following 
      print*, "Value of a is 10" 
   
   else if( a == 20 ) then
  
      ! if else if condition is true 
      print*, "Value of a is 20" 
  
   else if( a == 30 ) then
   
      ! if else if condition is true  
      print*, "Value of a is 30" 
  
   else
   
      ! if none of the conditions is true 
      print*, "None of the values is matching" 
      
   end if
   
   print*, "exact value of a is ", a
 
end program ifElseIfElseProg
```

```fortran
program datetime
    implicit none

    character(len = 8) :: dateinfo ! ccyymmdd
    character(len = 4) :: year
    character(len = 2) :: month, day

    character(len = 10) :: timeinfo ! hhmmss.sss
    character(len = 2)  :: hour, minute
    character(len = 6)  :: second

    call date_and_time(dateinfo, timeinfo)

    ! Extract date components
    year  = dateinfo(1:4)
    month = dateinfo(5:6)
    day   = dateinfo(7:8)

    ! Extract time components
    hour   = timeinfo(1:2)
    minute = timeinfo(3:4)
    second = timeinfo(5:10)

    ! Display results
    print *, 'dateinfo: ', dateinfo
    print *, 'timeinfo: ', timeinfo
    print *, "-----------------------"
    print *, "Current Date and Time (ISO 8601):"
    print *, "-----------------------"
    print *, "Date: ", year, "-", month, "-", day
    print *, "Time: ", hour, ":", minute, ":", second
end program datetime
```

```fortran
program zero
  implicit none
  real :: x_0, x, diff
  real, parameter :: tol = 1.0e-6
  x_0 = 1.0
  newton: do
    x = x_0 - ( x_0 - exp( - x_0 ) )/( 1 + exp( - x_0 ) )

    if ( x_0 /= 0.0 ) then
      diff = abs( ( x - x_0 )/( x_0 ) )
    else
      diff = abs( x - x_0 )
    end if
    x_0 = x
    if ( diff < tol )  then
      exit
    end if
  end do newton
  print '(a3,f8.5)', 'x =', x
end program zero
```

```fortran
! subroutine

program exproc
 implicit none
 integer, parameter :: n = 100
 real :: moy, max
 real, dimension (n) :: tab

 call random_number(tab)
 call sp (tab, moy, max)

 print '(a5,f8.5)','moy =', moy
 print '(a5,f8.5)', 'max =', max

contains
 subroutine sp(t, moy, max)
  implicit none
  real, dimension(:), intent(in) :: t
  real, intent(out) :: moy, max
  integer :: i, n
  n = size(t)
  max = t(1); moy = t(1)
  do i = 2, n
   if (t(i) > max) then
    max = t(i)
   end if
   moy = moy + t(i)
  end do
  moy = moy / n

 end subroutine sp
 
end program exproc
```
